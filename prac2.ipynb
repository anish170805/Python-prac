```python
from typing import List

class Solution:
    def runningSum(self, nums: List[int]) -> List[int]:
        result = []
        total = 0
        for num in nums:
            total += num
            result.append(total)
        return result

if __name__ == "__main__":
    sol = Solution()
    nums = [1, 2, 3, 4]
    print("Input:", nums)
    print("Running Sum:", sol.runningSum(nums))
```

```python
class Solution:
    def shuffle(self, nums, n):
        temp = [0] * (2 * n)

        i, j = 0, 0
        while i < 2 * n:
            temp[i] = nums[j]
            i += 2
            j += 1

        i, j = 1, n
        while i < 2 * n:
            temp[i] = nums[j]
            i += 2
            j += 1

        return temp

if __name__ == "__main__":
    sol = Solution()
    nums = [1,3,5,2,4,6]
    print("Input:", nums)
    print("Shuffled:", sol.shuffle(nums,len(nums)//2))
```

```python
from typing import List

class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        max_candies = max(candies)

        result = []
        for candy in candies:
            if candy + extraCandies >= max_candies:
                result.append(True)
            else:
                result.append(False)
        return result

if __name__ == "__main__":
    sol = Solution()
    candies = [2, 3, 5, 1, 3]
    extraCandies = 3
    print("Input:", candies, "Extra:", extraCandies)
    print("Output:", sol.kidsWithCandies(candies, extraCandies))
```
