### **Functions**

Screenshot 2025-12-06 221714.png

```python
def is_leap(year):
    leap = False

    if(year%4==0):
        if(year%100!=0):
            leap=True
        else:
            if(year%400==0):
                leap=True

    return leap

is_leap(1994)
```
